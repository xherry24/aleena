```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
numeric_cols = ['Unit price', 'Quantity', 'Total']

X = df[numeric_cols]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
inertia = []
K = range(1, 10)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.plot(K, inertia, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal K')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)

df[['Unit price','Quantity','Total','Cluster']].head()
```

```python
sns.scatterplot(data=df, x='Unit price', y='Total', hue='Cluster', palette='Set1')
plt.title("K-Means Clustering (2D)-22BDS0369")
plt.show()
```

```python
fig = px.scatter_3d(df, x='Unit price', y='Quantity', z='Total',
                    color='Cluster', symbol='Cluster',
                    color_discrete_sequence=px.colors.qualitative.Set1)
fig.update_layout(title="K-Means Clusters (3D)-22BDS0369")
fig.show()
```
